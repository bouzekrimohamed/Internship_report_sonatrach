```python
import random
import time

class Vehicle:
    def __init__(self, position, speed, destination):
        self.position = position
        self.speed = speed
        self.destination = destination

    def move(self):
        self.position -= self.speed

    def detect_and_share_obstacles(self, obstacle_position, obstacle_type):
        obstacle_distance = obstacle_position - self.position
        print(f"Distance de voiture {self.index} et la {obstacle_type} : {obstacle_distance} mètres")

    def detect_and_share_exit(self, exit):
        if exit is not None:
            distance_to_exit = abs(self.position - exit['position'])
            print(f"Distance de voiture {self.index} et la sortie {exit['destination']}: {distance_to_exit} mètres")
            if distance_to_exit < 20:
                print(f"(Vers {exit['destination']}, prendre la piste de secours {exit['lane']} vers {exit['destination']} à une distance de {distance_to_exit} mètres)")
        else:
            print(f"Distance de voiture {self.index} et la sortie {self.destination}: N/A")
            print(f"(Vers {self.destination}, pas de piste de secours disponible)")

    def change_speed(self):
        self.speed = random.randint(1, 5)

def choose_random_destination():
    return random.choice(['PARIS', 'MARSEILLE', 'ROUEN'])

def generate_random_obstacle():
    return random.randint(0, 100)  # Obstacle est situé à une position aléatoire sur l'autoroute

def generate_obstacle_type():
    return random.choice(['accident', 'manifestation', 'barrage_police'])  # Type d'obstacle aléatoire

def generate_exits():
    destinations = ['PARIS', 'MARSEILLE', 'ROUEN', 'LYON', 'TOULOUSE', 'NICE']
    num_exits = random.randint(1, 3)  # Nombre aléatoire de sorties
    exits = []
    for _ in range(num_exits):
        destination = random.choice(destinations)
        lane = random.randint(1, 5)  # Piste de secours aléatoire
        position = random.randint(0, 100)  # Position aléatoire sur l'autoroute
        exits.append({'destination': destination, 'lane': lane, 'position': position})
    return exits

def detect_and_share_obstacles(vehicle_data, obstacle_position, obstacle_type):
    for vehicle in vehicle_data:
        vehicle.detect_and_share_obstacles(obstacle_position, obstacle_type)

def detect_and_share_exits(vehicle_data, exits):
    for vehicle in vehicle_data:
        exit_for_vehicle = next((exit_info for exit_info in exits if exit_info['destination'] == vehicle.destination), None)
        vehicle.detect_and_share_exit(exit_for_vehicle)

def calculate_average_speed(vehicle_data):
    total_speed = sum(vehicle.speed for vehicle in vehicle_data)
    return total_speed / len(vehicle_data)

def evaluate_highway_status(vehicle_data, obstacle_threshold):
    distances = [vehicle.position for vehicle in vehicle_data]
    max_distance = max(distances)
    min_distance = min(distances)
    if max_distance - min_distance < obstacle_threshold:
        return "Fluide"
    else:
        return "Embouteillée"

def wait(seconds):
    time.sleep(seconds)

def main(num_updates):
    num_vehicles = 20  # Nombre de voitures
    vehicles = [Vehicle(random.randint(0, 100), random.randint(1, 5), choose_random_destination()) for _ in range(num_vehicles)]
    obstacle_threshold = 20

    print("Nombre de véhicules dans l'autoroute:", num_vehicles)
    print("Informations sur les véhicules:")
    for i, vehicle in enumerate(vehicles):
        vehicle.index = i + 1  # Ajout de l'indice de véhicule pour l'affichage
        print(f"Véhicule {vehicle.index}:")
        print("   Destination:", vehicle.destination)

    print("\nGénération de l'obstacle initial...")
    obstacle_position = generate_random_obstacle()
    obstacle_type = generate_obstacle_type()
    print(f"LE TYPE DE L'OBSTACLE DANS CETTE AUTOROUTE EST :: {obstacle_type}\n")

    print("Détection des obstacles...")
    detect_and_share_obstacles(vehicles, obstacle_position, obstacle_type)

    exits = generate_exits()

    for update in range(num_updates):
        print("\nMISE À JOUR ...")
        for vehicle in vehicles:
            vehicle.change_speed()  # Changer la vitesse de chaque véhicule
            vehicle.move()  # Déplacer le véhicule

        print("\nDétection des pistes de secours...")
        detect_and_share_exits(vehicles, exits)

        obstacle_position -= 5  # Déplacer l'obstacle vers les véhicules
        print("\nDétection des obstacles...")
        detect_and_share_obstacles(vehicles, obstacle_position, obstacle_type)

        print("\nVitesse moyenne de l'autoroute:", calculate_average_speed(vehicles), "m/s")
        print("État de l'autoroute:", evaluate_highway_status(vehicles, obstacle_threshold))

        wait(5)  # Attente de 5 secondes

if __name__ == "__main__":
    num_updates = 5  # Nombre de mises à jour
    main(num_updates)

```
